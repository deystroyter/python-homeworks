```python
import numpy as np
import pandas as pd
```

```python
UNRATED_MARK = -1
USERNAME = 'User 37'
SIM_COL = 'sim_col'
AVG_COL = 'avg_col'
KNN = 4
data = pd.read_csv('./csv/data.csv', index_col ='Unnamed: 0', delimiter=', ', engine='python')
data.head()
```

# Часть 1. 
**Для заданного пользователя (совпадает с вашим номером в таблице) рассчитать оценки для всех фильмов, которые он не оценил.**

```python
#Добавим столбец sim_userX с будущими оценками схожести
df_sim_userX = data.assign(sim_col = UNRATED_MARK)
#Добавим столбец со средними оценками
avg_user_rate = df_sim_userX.assign(avg_col = UNRATED_MARK)
```

```python
#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in df_sim_userX.drop(USERNAME).index:       
    temp = data.loc[[user, USERNAME]].transpose()
    temp = temp.loc[(temp[user] != UNRATED_MARK) &  (temp[USERNAME] != UNRATED_MARK)]
    temp_user = temp[user].to_numpy()
    temp_username = temp[USERNAME].to_numpy()
    
    sim = sum(temp_user * temp_username)/(np.sqrt(sum(temp_user*temp_user)) * np.sqrt(sum(temp_username * temp_username)))
    sim = round(sim, 3)
    df_sim_userX.loc[user, SIM_COL] = sim
```

```python
#Прогонимся по всем User'ам и вычислим для каждого среднюю оценку
for user in df_sim_userX.index:
    df_sim_userX.loc[user, AVG_COL] = round(np.average(data.loc[user][data.loc[user] != UNRATED_MARK]), 3)
    
df_sim_userX.head()
```

```python
#Это для ответа
answ = data.loc[USERNAME][data.loc[USERNAME] == UNRATED_MARK].to_frame()
print(answ)
```

```python
#Список непросмотренных фильмов
unRated_movies = list(data.loc[USERNAME][data.loc[USERNAME] == UNRATED_MARK].index)
```

```python
#Отсортируем строки по убыванию сходства
df_sim_userX = df_sim_userX.sort_values(by=SIM_COL, ascending = False)
df_sim_userX.head()
```

```python
#Для каждого непросмотренного фильма найдём KNN "ближайших соседей" с имеющейся оценкой
for movie in unRated_movies:
    temp = df_sim_userX.iloc[0:KNN][movie].to_frame()
    temp = temp.loc[temp.loc[:][movie] != UNRATED_MARK][movie]
    temp_sum = 0
    temp_sim_sum = 0
    
    #считаем суммы для формулы оценки
    for user in temp.index:   
        temp_sum += df_sim_userX.loc[user][SIM_COL] * (df_sim_userX.loc[user][movie] - df_sim_userX.loc[user][AVG_COL])
        temp_sim_sum += df_sim_userX.loc[user][SIM_COL]

    answ.loc[movie] = round(df_sim_userX.loc[USERNAME][AVG_COL] + temp_sum/temp_sim_sum, 3)
```

```python
print(answ)
```

```python
movie_names = pd.read_csv('./csv/movie_names.csv', delimiter=', ', engine='python')
context_day = pd.read_csv('./csv/context_day.csv', index_col=0, delimiter=', ', engine='python')
context_place = pd.read_csv('./csv/context_place.csv', index_col=0, delimiter=', ', engine='python')
```

```python
#Добавим оценки схожести для context_place
m_context_place = context_place.assign(sim_col = str(UNRATED_MARK))

#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in m_context_place.drop(USERNAME).index:
    temp = m_context_place.loc[[user, USERNAME]].transpose()
    temp = temp.loc[(temp[user] != str(UNRATED_MARK)) &  (temp[USERNAME] != str(UNRATED_MARK))]
    temp_user = temp[user].to_numpy()
    temp_username = temp[USERNAME].to_numpy()
    sim = round(sum(temp_user == temp_username) / len(temp_user), 3)
    m_context_place.loc[user, SIM_COL] = sim
m_context_place.head()
```

```python
#Добавим оценки схожести для context_day тут для решения конкретной задачи, будет удобно
#заменить Субботу и Воскресенье на одинаковое значение, например 'End'
#по такому же принципу заменить будни, пусть будет 'Day'
m_context_day = context_day.assign(sim_col = str(UNRATED_MARK))
m_context_day = m_context_day.replace(['Mon','Tue','Wed','Thu','Fri'], 'Day')
m_context_day = m_context_day.replace(['Sat', 'Sun'], 'End')
```

```python
#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in m_context_day.drop(USERNAME).index:
    temp = m_context_day.loc[[user, USERNAME]].transpose()
    temp = temp.loc[(temp[user] != str(UNRATED_MARK)) &  (temp[USERNAME] != str(UNRATED_MARK))]
    temp_user = temp[user].to_numpy()
    temp_username = temp[USERNAME].to_numpy()
    sim = round(sum(temp_user == temp_username) / len(temp_user), 3)
    m_context_day.loc[user, SIM_COL] = sim
m_context_day.head()
```

```python
#Перемножим все схожести
for user in df_sim_userX.drop(USERNAME).index:       
    df_sim_userX.loc[user, SIM_COL] = round(df_sim_userX.loc[user, SIM_COL] * m_context_place.loc[user, SIM_COL] * m_context_day.loc[user, SIM_COL], 3)

#Отсортируем строки по убыванию сходства
df_sim_userX = df_sim_userX.sort_values(by=SIM_COL, ascending = False)
df_sim_userX.head()
```

```python
#Это для ответа.
answ2 = answ.copy()

#Будем плюсовать Наш рейтинг, [1 в выходной | 0 в будний], [1 для дома | 0.5 в гостях | 0 в кинотеатре]
#Для каждого непросмотренного фильма найдём "ближайших соседей" с имеющейся оценкой
for movie in unRated_movies:
    temp_sum_rate = 0
    temp_data = df_sim_userX.drop(USERNAME)
    temp_data = df_sim_userX.loc[:][df_sim_userX.loc[:][movie] != UNRATED_MARK]

    for user in temp_data.index.values[0:10]:
        #учтём оценки "контекстных" пользователей
        temp_sum_rate += temp_data.loc[user, movie] * temp_data.loc[user, SIM_COL]
        
        if m_context_day.loc[user, movie] == 'End':
            answ2.loc[movie] += 1 * temp_data.loc[user, SIM_COL] 
        if m_context_place.loc[user, movie] == 'h':
            answ2.loc[movie] += 1 * temp_data.loc[user, SIM_COL]
        elif m_context_place.loc[user, movie] == 'v':
            answ2.loc[movie] += 0.5 * temp_data.loc[user, SIM_COL]
    answ2.loc[movie] += temp_sum_rate / KNN
    answ2.loc[movie] = round(answ2.loc[movie], 3)
            
answ2 = answ2.sort_values(by=USERNAME, ascending = False)
print(answ2)
print(answ2.index[0])
```

```python
from SPARQLWrapper import SPARQLWrapper, JSON
import requests
movie_names = pd.read_csv('./csv/movie_names.csv',index_col = 0, header = None, delimiter=', ', engine='python')
temp = answ2.index[0]
```

```python
#"Вытянем" имя фильма из movie_names
film_name = movie_names.loc[temp][1]
print(film_name)
```

```python
#Найдём все Entities с "The Shawshank Redemption" в названии
API_ENDPOINT = "https://www.wikidata.org/w/api.php"

params = {
    'action' : 'wbsearchentities',
    'format' : 'json',
    'language' : 'en',
    'search': film_name
}

request_result = requests.get(API_ENDPOINT, params = params).json()['search']
```

```python
#Из полученных страничек, отберём те, где в описании есть слово 'film'
#(так как к этому фильму ещё идёт одноимённый музыкальный альбом)
for item in request_result:
    if 'film' in item['description'].split(' '):
        film_wiki_id = item['id']
```

```python
film_wiki_id
```

```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
```

```python
sparql_query = '''
SELECT DISTINCT ?film ?filmLabel ?recom_year
WHERE 
{  
    { 
      SELECT DISTINCT (MIN(YEAR(?recom_date)) as ?recom_year)
      WHERE
     {
        ?film wdt:P31 wd:Q11424 filter(?film = wd:'''+film_wiki_id+''').
        ?film wdt:P577 ?recom_date;
                    
      }
    }

  
  ?film wdt:P31  wd:Q11424 .
  ?film wdt:P577 ?date  filter(BOUND(?date) && YEAR(?date) = ?recom_year) .

  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
  
}

'''

sparql.setQuery(sparql_query)
sparql.setReturnFormat(JSON)
results = sparql.query()
```

```python
results = results.convert()
```

```python
df_results = pd.json_normalize(results['results']['bindings'])
df_results.head(5)
```
