```python
import numpy as np
import pandas as pd
```

```python
unRatedMark = -1
username = 'User 37'
sim_col = 'sim_col'
knn = 4
data = pd.read_csv('./csv/data.csv', index_col ='Unnamed: 0', delimiter=', ', engine='python')
data.head()
```

# Часть 1. 
**Для заданного пользователя (совпадает с вашим номером в таблице) рассчитать оценки для всех фильмов, которые он не оценил.**

```python
#Добавим столбец sim_userX с будущими оценками схожести
df_sim_userX = data.assign(sim_col = unRatedMark)
```

```python
#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in df_sim_userX.drop(username).index:       
    temp = df_sim_userX.loc[[user, username]].transpose()
    temp = temp.loc[(temp[user] != unRatedMark) &  (temp[username] != unRatedMark)]
    temp_user = temp[user].to_numpy()
    temp_username = temp[username].to_numpy()
    
    sim = sum(temp_user * temp_username)/(np.sqrt(sum(temp_user*temp_user)) * np.sqrt(sum(temp_username * temp_username)))
    sim = round(sim, 3)
    df_sim_userX.loc[user, sim_col] = sim


df_sim_userX.head(40)
```

```python
#Это для ответа
answ = data.loc[username][data.loc[username] == unRatedMark].to_frame()
print(answ)
```

```python
#Список непросмотренных фильмов
unRated_movies = list(data.loc[username][data.loc[username] == unRatedMark].index)
```

```python
#Отсортируем строки по убыванию сходства
df_sim_userX = df_sim_userX.sort_values(by=sim_col, ascending = False)
df_sim_userX.head()
```

```python
#Для каждого непросмотренного фильма найдём "ближайших соседей" с имеющейся оценкой
for movie in unRated_movies:
    temp = df_sim_userX.loc[df_sim_userX.loc[:][movie] != unRatedMark][movie]
    temp = np.average(temp[0:knn].to_numpy())
    answ.loc[movie] = temp
```

```python
print(answ)
```

# **Часть 2.**
**Порекомендовать заданному пользователю 1
фильм, который он посмотрел бы в выходной дома,
если такой фильм есть в данных.**

```python
movie_names = pd.read_csv('./csv/movie_names.csv', delimiter=', ', engine='python')
context_day = pd.read_csv('./csv/context_day.csv', index_col=0, delimiter=', ', engine='python')
context_place = pd.read_csv('./csv/context_place.csv', index_col=0, delimiter=', ', engine='python')
```

> Я не придумал ничего лучше, чем просто посчитать для каждого контекста схожесть как кол-во совпадающих на общее кол-во
имеющихся

> Посчитать общую схожесть, учитывая схожесть оценок за фильмы

> И "высказать решение" большинства ближайших соседей по каждому контексту

```python
#Добавим оценки схожести для context_place
m_context_place = context_place.assign(sim_col = str(unRatedMark))

#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in m_context_place.drop(username).index:
    temp = m_context_place.loc[[user, username]].transpose()
    temp = temp.loc[(temp[user] != str(unRatedMark)) &  (temp[username] != str(unRatedMark))]
    temp_user = temp[user].to_numpy()
    temp_username = temp[username].to_numpy()
    sim = round(sum(temp_user == temp_username) / len(temp_user), 3)
    m_context_place.loc[user, sim_col] = sim
m_context_place.head()
```

```python
#Добавим оценки схожести для context_day тут для решения конкретной задачи, будет удобно
#заменить Субботу и Воскресенье на одинаковое значение, например 'End'
#по такому же принципу заменить будни, пусть будет 'Day'
m_context_day = context_day.assign(sim_col = str(unRatedMark))
m_context_day = m_context_day.replace(['Mon','Tue','Wed','Thu','Fri'], 'Day')
m_context_day = m_context_day.replace(['Sat', 'Sun'], 'End')
```

```python
#Прогонимся по всем User'ам и вычислим для каждого сходство с искомым
for user in m_context_day.drop(username).index:
    temp = m_context_day.loc[[user, username]].transpose()
    temp = temp.loc[(temp[user] != str(unRatedMark)) &  (temp[username] != str(unRatedMark))]
    temp_user = temp[user].to_numpy()
    temp_username = temp[username].to_numpy()
    sim = round(sum(temp_user == temp_username) / len(temp_user), 3)
    m_context_day.loc[user, sim_col] = sim
m_context_day.head()
```

```python
#Перемножим все схожести
for user in df_sim_userX.drop(username).index:       
    df_sim_userX.loc[user, sim_col] = round(df_sim_userX.loc[user, sim_col] * m_context_place.loc[user, sim_col] * m_context_day.loc[user, sim_col], 3)

#Отсортируем строки по убыванию сходства
df_sim_userX = df_sim_userX.sort_values(by=sim_col, ascending = False)
df_sim_userX.head()
```

```python
#Это для ответа.
answ2 = answ.copy()

#Будем плюсовать Наш рейтинг, [1 в выходной | 0 в будний], [1 для дома | 0.5 в гостях | 0 в кинотеатре]

#Для каждого непросмотренного фильма найдём "ближайших соседей" с имеющейся оценкой
for movie in unRated_movies:
    for user in df_sim_userX.drop(username).index.values[0:knn]:
        if m_context_day.loc[user, movie] == 'End':
            answ2.loc[movie] += 1
        if m_context_place.loc[user, movie] == 'h':
            answ2.loc[movie] += 1
        elif m_context_place.loc[user, movie] == 'v':
            answ2.loc[movie] += 0.5
            
answ2 = answ2.sort_values(by=username, ascending = False)
print(answ2.index[0])
```

```python
#Запись в json
result = {
    'user': username,
    '1': dict(zip(answ.index.tolist(), answ.values.tolist())),
    '2': dict(zip(answ2.index.tolist(), answ2.values[0].tolist()))
}
print(result)
import json
with open('result.json', 'w') as fp:
    json.dump(result, fp, indent=4)
```
